```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
%pylab inline

objects = ('Linear', 'Sigmoid', 'RBF', 'poly', 'Linear_Elastic')
y_pos = np.arange(len(objects))
performance = [95.45,79.80,92.81,95.53,98.47]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('Graph Shows Accuracy Varies as Kernal Changes')

plt.show()
```
